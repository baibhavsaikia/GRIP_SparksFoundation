### Import the necessary libraries

```python
import numpy as np                                         
import pandas as pd                                        
import matplotlib.pyplot as plt                            
import sklearn.datasets as datasets                        
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

### Reading data and its features

```python
iris=datasets.load_iris()                                      #Loading the iris dataset
```

```python
X = pd.DataFrame(iris.data, columns=iris.feature_names)        #Forming the iris dataframe
```

```python
X.head()                                                       #prints the first 5 rows
```

```python
X.tail()                                                        #prints the last 5 rows
```

```python
X.describe()                                                     #prints the descriptive statistics of data
```

```python
X.shape                                                         #prints the dimensions(no.of rows, no.of columns)
```

```python
X.info()                                                        #Summarize the dataset
```

```python
Y=iris.target
Y
```

### Defining the Decision Tree Algorithm

```python
dt=DecisionTreeClassifier()                        
dt.fit(X,Y)
```

### Visualization of the decision tree

```python
featName=['sepal length ( in cm)','sepal width (in cm)','petal length (in cm)','petal width (in cm)']
clsName=['Iris-Setosa','Iris-Versicolor','Iris-virginica']

fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (10,6), dpi = 500)

tree.plot_tree(dt, feature_names = featName, class_names = clsName, filled = True);
```
